```python
import numpy as np


from learning_algorithms.network import Network
from pandas import read_csv, DataFrame, Series


data = read_csv('train.csv')

#data.pivot_table('PassengerId', 'Pclass', 'Survived', 'count').plot(kind='bar', stacked=True)


data.drop('Cabin', 1, inplace = True)
data.drop('Ticket', 1, inplace = True)
data.drop('Name', 1, inplace = True)
data.drop('PassengerId', 1, inplace = True)

#print(data['Sex'])

for i in range(len(data['Sex'])):
    if(data['Sex'][i]=='male'):
        data['Sex'][i] = 1
    else:
        data['Sex'][i] = 0
    if data['Embarked'][i] == 'C':
        data['Embarked'][i] = 3
    elif data['Embarked'][i] == 'S':
        data['Embarked'][i] = 2
    elif data['Embarked'][i] == 'Q':
        data['Embarked'][i] = 1
    else:
        data['Embarked'][i] = 2
    

print(data['Age'].mean())
```

```python
print(data)
```

```python
import math
for i in range(len(data['Age'])):
    if math.isnan(data['Age'][i]):
        data['Age'][i] = data['Age'].mean()
print(data)
```

```python
data.isnull().any()
```

```python
#данные не пустые все пустые возроста заменил на среднее, может это и плохое решение
#стоило посмотреть как сказалось незнание возраста на их выживаемость
#возможно они погибли и поэтому про них неизвестен возраст
#пронормируем данные по возросту и цене т.к они на порядок отличаются от остальных
means_age = data['Age'].mean(axis = 0)
stds_age = data['Age'].std(axis=0)
```

```python
print(means_age)
print(stds_age)
```

```python
data['Age'] = (data['Age'] - means_age) / stds_age
```

```python
means_fare = data['Fare'].mean(axis = 0)
stds_fare = data['Fare'].std(axis=0)
data['Fare'] = (data['Fare'] - means_fare)/ stds_fare
print(data)
```

```python
print(means_fare)
print(stds_fare)
```

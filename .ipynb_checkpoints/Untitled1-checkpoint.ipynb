```python
import numpy as np
import math

from learning_algorithms.network import Network
from pandas import read_csv, DataFrame, Series


data = read_csv('train.csv')

test = read_csv('test.csv')

#готовим данные (избавляемся от лишних, заполняем пропущенные, нормализуем)

data.drop('Cabin', 1, inplace = True)
data.drop('Ticket', 1, inplace = True)
data.drop('Name', 1, inplace = True)
data.drop('PassengerId', 1, inplace = True)

test.drop('Cabin', 1, inplace = True)
test.drop('Ticket', 1, inplace = True)
test.drop('Name', 1, inplace = True)
test.drop('PassengerId', 1, inplace = True)

#переводим буквенные значения в цифровые
for i in range(len(data['Sex'])):
    if(data['Sex'][i]=='male'):
        data['Sex'][i] = 1
    else:
        data['Sex'][i] = 0
    if data['Embarked'][i] == 'C':
        data['Embarked'][i] = 3
    elif data['Embarked'][i] == 'S':
        data['Embarked'][i] = 2
    elif data['Embarked'][i] == 'Q':
        data['Embarked'][i] = 1
    else:
        data['Embarked'][i] = 2
    


        
#переводим буквенные значения в цифровые
for i in range(len(test['Sex'])):
    if(test['Sex'][i]=='male'):
        test['Sex'][i] = 1
    else:
        test['Sex'][i] = 0
    if (test['Embarked'][i] == 'C'):
        test['Embarked'][i] = 3
    elif (test['Embarked'][i] == 'S'):
        test['Embarked'][i] = 2
    elif (test['Embarked'][i] == 'Q'):
        test['Embarked'][i] = 1
    else:
        test['Embarked'][i] = 2

#нормализация        
means_age = data['Age'].mean(axis = 0)
stds_age = data['Age'].std(axis=0)

means_fare = data['Fare'].mean(axis = 0)
stds_fare = data['Fare'].std(axis=0)

#заполняем пустые строки    
for i in range(len(data['Age'])):
    if math.isnan(data['Age'][i]):
        data['Age'][i] = means_age

#заполняем пустые строки 
for i in range(len(test['Age'])):
    if math.isnan(test['Age'][i]):
        test['Age'][i] = means_age

#нормализация
data['Age'] = (data['Age'] - means_age) / stds_age
data['Fare'] = (data['Fare'] - means_fare)/ stds_fare

test['Age'] = (test['Age'] - means_age)/ stds_age
test['Fare'] = (test['Fare'] - means_fare)/ stds_fare
print(data)
print(test)


```

```python
#заполняем пустые строки (может это и не лучший способ заполнения, но пока так)
test_fare_means = test['Fare'].mean(axis = 0)
for i in range(len(test['Fare'])):
    if math.isnan(test['Fare'][i]):
        test['Fare'][i] = test_fare_means
test.isnull().any()
```

```python
#преобразуем данные в массив понимаемый нейросетью
l = len(data['Sex'])
train = np.zeros((l, 8))
for i in range(l):
    train[i][0] = data['Pclass'][i]
    train[i][1] = data['Sex'][i]
    train[i][2] = data['Age'][i]
    train[i][3] = data['SibSp'][i]
    train[i][4] = data['Parch'][i]
    train[i][5] = data['Fare'][i]
    train[i][6] = data['Embarked'][i]
    train[i][7] = data['Survived'][i]
   
p = len(test['Sex'])
re_test = np.zeros((p,7)) 
for i in range(p):
    re_test[i][0] = test['Pclass'][i]
    re_test[i][1] = test['Sex'][i]
    re_test[i][2] = test['Age'][i]
    re_test[i][3] = test['SibSp'][i]
    re_test[i][4] = test['Parch'][i]
    re_test[i][5] = test['Fare'][i]
    re_test[i][6] = test['Embarked'][i]
re_test = [ex.reshape((7,1)) for ex in re_test]
print(re_test)
```

```python
train = [(d[:7][:, np.newaxis], np.eye(2, 1, k=-int(d[-1]))) for d in train]
print(train)
```

```python
#создаём нейросеть и обучаем её
input_count  = 7  
hidden_count = 10  
output_count = 2
nn = Network([input_count, hidden_count, output_count], l2 = 0.1)
nn.SGD(train, 100, mini_batch_size=5, eta=0.25)
```

```python
prediction = list()
for ex in re_test:
    prediction.append(nn.feedforward(ex))
print(prediction)
```
